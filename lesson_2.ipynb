С помощью функции `linspace` библиотеки `Nump`y создаем массив `x` из 51 числа от -3 до 3 включительно.

```python
x = np.linspace(-3, 3, 51)
x
```

Создаем массивы `y1`, `y2`, `y3`, `y4` по следующим формулам:
`y1 = x**2`
`y2 = 2 * x + 0.5`
`y3 = -3 * x - 1.5`
`y4 = sin(x)`

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

Используя функцию `subplots` модуля `matplotlib.pyplot`, создаем объект `matplotlib.figure`.`Figure` с названием `fig` и массив объектов `Axes` под названием `ax`,причем так, чтобы у нас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов.
В каждом графике массив `x` используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте `y1`,в правом верхнем - `y2`, в левом нижнем - `y3`, в правом нижнем - `y4`.

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
```

Задаем название графикам: 'График y1', 'График y2' и т.д.

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_title('График y1')

ax2.plot(x, y2)
ax2.set_title('График y2')

ax3.plot(x, y3)
ax3.set_title('График y3')

ax4.plot(x, y4)
ax4.set_title('График y4')
```

Для графика в левом верхнем углу установим границы по оси `x` от -5 до 5.

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2')

ax3.plot(x, y3)
ax3.set_title('График y3')

ax4.plot(x, y4)
ax4.set_title('График y4')
```

Установим размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками зададим 0.3 доли от среднего значения осей графика.

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2')

ax3.plot(x, y3)
ax3.set_title('График y3')

ax4.plot(x, y4)
ax4.set_title('График y4')
```

Получим матрицу `B`, умножив матрицу `A` на матрицу `At` с помощью матричного умножения. Какой размер имеет матрица `B`? Получится ли вычислить обратную матрицу для матрицы `B` и почему?

```python
a = np.arange(0, 12)
A = a.reshape(4, 3)
At = A.T
B = A.dot(At)
print('Размер матрицы В:', B.shape)
# Обратную матрицу для матрицы `В` не получится вычислить, т.к. матрица В - квадратная, но ее определитель равен 0.
print('Определитель матрицы В:', np.linalg.det(B))
```

11.	Создадим одномерный массив `c`, составленный из последовательности 16-ти случайных равномерно распределенных целых чисел от 0 до 16 невключительно.

12.	Поменяем его форму так, чтобы получилась квадратная матрица `C`. Получим матрицу `D`, поэлементно прибавив матрицу `B` из предыдущего вопроса к матрице `C`, умноженной на 10. Вычислим определитель, ранг и обратную матрицу `D_inv` для `D`.

```python
C = c.reshape(4, 4)
D = C * 10 + B

print('Определитель матрицы D:', np.linalg.det(D))
print('Ранг матрицы D:', np.linalg.matrix_rank(D))
D_inv = np.linalg.inv(D)
print('Обратная матрица D_inv =', D_inv)
```

С помощью функции `numpy.where`, используя матрицу `D_inv` в качестве маски, а матрицы `B` и `C` - в качестве источников данных, получим матрицу `E` размером 4x4.  Элементы матрицы `E`, для которых соответствующий элемент матрицы `D_inv` равен 1, равны соответствующему элементу матрицы `B`, а элементы матрицы `E`, для которых соответствующий элемент матрицы `D_inv` равен 0, равны соответствующему элементу матрицы `C`.

```python
D_inv[D_inv < 0.0] = 0
D_inv[D_inv > 0.0] = 1
print(D_inv)

E = D_inv.copy()
print(E)

E[D_inv == 0.0] = C[np.where(D_inv == 0)]
E[D_inv == 1.0] = B[np.where(D_inv == 1)]

print(C)
print(B)
print(E)
```
